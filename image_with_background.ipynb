```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def change_back(background, img):
  """ Una funcin que reemplaza el fondo negro de las imgenes por otra imagen.
    Acepta dos argumentos: la imagen de fondo (background) y la propia imagen (img).
    La funcin comenzar desde la esquina superior izquierda de la imagen.
    Buscar todos los pxeles negros (con el valor 0) y los reemplazar por pxeles de la imagen de fondo."""
  x, y = 0, 0
  background = cv2.resize(background, (img.shape[1], img.shape[0]), interpolation = cv2.INTER_AREA)
  res = np.copy(background)
  place = res[y: y + img.shape[0], x: x + img.shape[1]]
  a = img[..., 3:].repeat(3, axis=2).astype('uint16')
  place[...] = (place.astype('uint16') * (255 - a) // 255) + img[..., :3].astype('uint16') * a // 255
  return res

# Load the background image
background = cv2.imread("coloes.jpg", cv2.IMREAD_UNCHANGED)

# Load the image with transparent background
img = cv2.imread("gecko_sin_fondo.jpeg", cv2.IMREAD_UNCHANGED)

# Call the function and store the result
result = change_back(background, img)


# Save the resulting image
cv2.imwrite("gecko_con_fondo_texto.jpg", result)

# Display the resulting image with text
cv2_imshow(result)



```
